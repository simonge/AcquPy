# AcquDetector module
The detector module, holds all of the information on the detectors and how to convert the raw adc values into relevent detector signals.

Detector information has been converted into the JSON file format. Currently with one file for each measurement parameter (e.g. NaI Energy, NaI Time, BaF Energy, BaF SG Energy, BaF Time, Tagger Time)

The JSON format:
    
    detector   : Unique ID String (e.g. NaI or CB)
    calEq      : python equation which applies calibration to adcs
    channels   : array of channels
        channel    : channel number
        adc        : adc number relating to measurement
        raw        : empty python array raw adc value will be entered into
        value      : empty python array calibrated value will be entered into
        ref        : array number of reference channel if required by calEq
        neighbours : necessary if being interogated as a graph
        [consts]   : any values required by calibration or analysis
    references : array of required reference adcs
        adc        : adc number relating to reference channel
        raw        : empty python array raw adc value will be entered into
       
Acqu setup files can be easily converted using python scripts.

The calEq which allows the detector to self calibrate is possibly not very user friendly.

# Calibrating the adc values

Calling Calibrate on the raw adc array filters the array into the different detectors loaded and applies the calibration equation returning a python dictionary containing a numpy array with all of each channels parameters.

```python
import ROOT
import numpy as np
from rootpy.plotting import histogram, Hist2D, Hist, Canvas

taggerChannels = aqdet.Channels['tagger']
taggerchan     = Hist(taggerChannels,0,taggerChannels,name='taggerChan')
CBHitsE = Hist2D(200,0,200,200,0,2000,name='CBHitsE')
CBChanE = Hist2D(aqdet.Channels['CB'],0,aqdet.Channels['CB'],100,0,1000,name='CBEChan')

def plotCalCB():
    data = aqdet.Calibrate(aq.adcArray)    
    CBHitsE.Fill(data['CB']['value'].size,np.sum(data['CB']['value']))
    for dat in data['CB']:
        CBChanE.Fill(dat['channel'],dat['value'])
    for dat in data['tagger']:
        taggerchan.Fill(dat['channel'])
        
    if(aq.eventNo%1000==0):
        print("Events Processed: ",aq.eventNo)
    
aq.runFunction(plotCalCB,0,5000)

ROOT.enableJSVis()
c2 = ROOT.TCanvas("my3","The Canvas Title",800,600)
taggerchan.Draw("colz")
c2.Draw()
c0 = ROOT.TCanvas("my","The Canvas Title",800,600)
CBHitsE.Draw("colz")
c0.Draw()
c1 = ROOT.TCanvas("my2","The Canvas Title",800,600)
CBChanE.Draw("colz")
c1.Draw()
```

# Plotting detector graphs

The plotly python package has some very nice plotting capabilities. The detectors verteces and neighbour edges can easily be plotted in 3D to examine their structures. Here we see my TAPS file has some problems!

```python
import plotly.offline as py
import plotly.graph_objs as go
py.init_notebook_mode()

# CB plot
aqdet.detgraphs['CB'].vs['label'] = aqdet.detgraphs['CB'].vs['channel']
FullCBnodes, FullCBedges     = aqdet.GraphPlot(aqdet.detgraphs['CB'])

# TAPS plot
aqdet.detgraphs['TAPS'].vs['label'] = aqdet.detgraphs['TAPS'].vs['channel']
FullTAPSnodes, FullTAPSedges = aqdet.GraphPlot(aqdet.detgraphs['TAPS'])

layout = go.Layout(
    title = 'title',
    width=1000,
    height=1000
)

figCB   = go.Figure(data=[FullCBnodes,FullCBedges],layout=layout)

figTAPS = go.Figure(data=[FullTAPSnodes,FullTAPSedges],layout=layout)

py.iplot(figCB)
py.iplot(figTAPS)
```

# Drawing detector hits

The way the detector and data is output makes it easy to draw the real data (as a sub-graph of the detector) ontop of the defined detector geometries. Here **CalGraph** is called instead of **calibrate** returning a graph for each event rather than an array. The graph package **igraph** has clustering algorithms but here each non-joined set is shown in a different colour.

```python
import math

def DrawDetectorHits():
    print('number',aq.eventNo)
    #print(A2G4.signals)
    
    data = aqdet.CalGraph(aq.adcArray)
            
    plots  = [aqdet.GraphPlot(plot,1)[0] for plot in data['CB'].decompose()]
    plots += [aqdet.GraphPlot(plot,1)[0] for plot in data['TAPS'].decompose()]
    plots += [aqdet.GraphPlot(plot,1)[1] for plot in data['CB'].decompose()]
    plots += [aqdet.GraphPlot(plot,1)[1] for plot in data['TAPS'].decompose()]
    
    fig = go.Figure(data=[FullCBedges,FullTAPSedges,FullCBnodes,FullTAPSnodes]+plots,layout=layout)

    py.iplot(fig)
    

    
    
    
```

```python
aqdet.LoadDetectors(['aux/CrystalBallEnergy.json','aux/TAPSEnergy.json'])
maxEvnts = 28
aq.runFunction(DrawDetectorHits,25,maxEvnts)
```
